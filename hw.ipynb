```python
import numpy as np
import matplotlib.pyplot as plt

# Define the x range for the functions
x1 = np.linspace(-3 * np.pi, 3 * np.pi, 10001)
x2 = np.linspace(-10, 10, 10001)

# Define the first function f1(x)
f1 = np.where(
    np.sin(x1) + np.cos(x1) >= 0,
    np.maximum(np.sin(x1), np.cos(x1)),
    np.minimum(np.sin(x1), np.cos(x1))
)

# Define the second function f2(x)
f2 = np.where(
    np.isclose(x2, np.round(x2)),
    -x2 / 10,
    np.sin(1 / np.where(x2 != 0, x2, np.nan))  # Avoid division by zero
)
# Plotting the functions
plt.figure(figsize=(10, 5))

# Scatter plot for f1(x)
plt.subplot(1, 2, 1)
plt.scatter(x1, f1, s=1)
plt.title(r"$f_1(x)$")
plt.xlabel(r"$x$")
plt.ylabel(r"$f_1(x)$")

# Scatter plot for f2(x)
plt.subplot(1, 2, 2)
plt.scatter(x2, f2, s=1)
plt.title(r"$f_2(x)$")
plt.xlabel(r"$x$")
plt.ylabel(r"$f_2(x)$")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function to perform the polygon iteration
def plot_polygon(n, i):
    # Step 1: Initialize n random (x, y) coordinates
    points = np.random.rand(n, 2)  # Generate n random points in 2D
    
    # Step 2: Perform the i iterations
    for _ in range(i):
        new_points = np.zeros_like(points)  # Array to store new coordinates
        # Compute the midpoints for each pair of consecutive points
        for i in range(len(points) - 1):
            new_points[i] = (points[i] + points[i + 1]) / 2
        # Last point is the midpoint between the last and first points
        new_points[-1] = (points[-1] + points[0]) / 2
        points = new_points  # Update points for the next iteration
    return points

# Test with n=50 and i=1000
points = plot_polygon(50, 1000)

# Step 3: Plot the points and lines
plt.figure(figsize=(6, 6))
plt.plot(points[:, 0], points[:, 1], 'o:', color="black", markersize=5, markerfacecolor="#0000FF", markeredgecolor="#000000", linewidth=1)
plt.axis('square')
plt.axis('off')
plt.show()
```

```python
import numpy as np

# Define the function H to compute the Hessian matrix
def H(*args):
    n = len(args)  # Determine the number of variables
    x = np.array(args)  # Convert inputs to a NumPy array for easy broadcasting

    # Compute cos(x) and sin(x) for all inputs
    cos_x = np.cos(x)
    sin_x = np.sin(x)
    
    # Compute the function value f(x) = cos(x1) * cos(x2) * ... * cos(xn)
    f_x = np.prod(cos_x)
    
    # Diagonal elements of the Hessian (second partial derivatives w.r.t. the same variable)
    diag_elements = -f_x * (sin_x / cos_x) ** 2 + (f_x / cos_x ** 2)

    # Off-diagonal elements of the Hessian (mixed second partial derivatives)
    off_diag_elements = f_x * (sin_x / cos_x)[:, None] * (sin_x / cos_x)[None, :]
    
    # Construct the Hessian matrix
    H_matrix = np.diag(diag_elements) + off_diag_elements
    return H_matrix

# Test the function with an example input
H_matrix = H(np.pi/4, np.pi/4, np.pi/4)
H_matrix
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the grid size
grid_size = (200, 200)

# Define growth parameters
min_growth = 0.01
max_growth = 0.04
max_height = 5
yield_threshold = 2

# Function to simulate crop growth for a specific harvest interval
def simulate_growth(harvest_interval, days):
    crop_heights = np.zeros(grid_size)  # Initialize crop heights
    total_yield = 0  # Accumulated yield at the end of each harvest interval
    yield_per_planted_day_list = []  # Store yield per planted day for each interval

    for day in range(1, days + 1):
        # Step 1: Daily growth
        daily_growth = np.random.uniform(min_growth, max_growth, grid_size)
        crop_heights = np.minimum(crop_heights + daily_growth, max_height)

        # Step 2: Only calculate yield on harvest day
        if day % harvest_interval == 0:
            # Calculate yield based on crop heights on the harvest day
            yield_today = np.sum(np.maximum(0, np.floor(crop_heights) - yield_threshold + 1))
            
            # Calculate yield per planted day for this interval
            yield_per_planted_day = yield_today / harvest_interval
            yield_per_planted_day_list.append(yield_per_planted_day)
            
            # Reset crop heights for the next growth cycle
            crop_heights = np.zeros(grid_size)
            total_yield = 0  # Reset yield after harvest

    return np.mean(yield_per_planted_day_list)

# Simulation parameters
days = 1000  # Total simulation days
harvest_intervals = range(100, 300)  # Testing intervals from 50 to 500 days

# Calculate average yield per planted day for each interval
avg_yields = []
for interval in harvest_intervals:
    avg_yield = simulate_growth(interval, days)
    avg_yields.append(avg_yield)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(harvest_intervals, avg_yields, marker='o')
plt.xlabel("Harvest Interval (days)")
plt.ylabel("Average Yield per Planted Day")
plt.title("Average Yield per Planted Day vs Harvest Interval")
plt.show()

# Determine optimal interval and corresponding yield per planted day
optimal_index = np.argmax(avg_yields)
optimal_interval = harvest_intervals[optimal_index]
optimal_yield = avg_yields[optimal_index]
print(f"Optimal harvest interval: {optimal_interval} days")
print(f"Corresponding yield per planted day: {optimal_yield}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import convolve

# Set the grid size
grid_size = (200, 200)

# Define growth parameters
min_growth = 0.01
max_growth = 0.04
max_height = 5
yield_threshold = 2

# Convolution kernel to detect neighbors for shading effect
kernel = np.array([[1, 1, 1], [1, 0, 1], [1, 1, 1]])

# Function to simulate crop growth with shading effect for a specific harvest interval
def simulate_growth_with_shading(harvest_interval, days):
    crop_heights = np.zeros(grid_size)  # Initialize crop heights
    yield_per_planted_day_list = []  # Store yield per planted day for each interval

    # Pre-generate random growth values for all days to reduce calls to np.random.uniform
    daily_growth_values = np.random.uniform(min_growth, max_growth, (days,) + grid_size)

    for day in range(1, days + 1):
        # Identify positions with height exactly 5 (shade sources)
        shade_sources = (crop_heights == max_height)

        # Check if any shade sources are present
        if np.any(shade_sources):
            # Calculate shaded positions only when there are shade sources
            shaded_positions = convolve(shade_sources.astype(int), kernel, mode='constant') > 0
            daily_growth = daily_growth_values[day - 1] * np.where(shaded_positions, 0.125, 1.0)
        else:
            daily_growth = daily_growth_values[day - 1]

        # Update crop heights, ensuring they do not exceed max height
        crop_heights = np.minimum(crop_heights + daily_growth, max_height)

        # Step 2: Only calculate yield on harvest day
        if day % harvest_interval == 0:
            # Calculate yield based on crop heights on the harvest day
            yield_today = np.sum(np.maximum(0, (crop_heights // 1) - yield_threshold + 1))
            
            # Calculate yield per planted day for this interval
            yield_per_planted_day = yield_today / harvest_interval
            yield_per_planted_day_list.append(yield_per_planted_day)
            
            # Reset crop heights for the next growth cycle
            crop_heights = np.zeros(grid_size)

    return np.mean(yield_per_planted_day_list)

# Simulation parameters
days = 1000  # Total simulation days
harvest_intervals = range(100, 300)  # Testing intervals from 50 to 500 days in steps of 10

# Calculate average yield per planted day for each interval with shading
avg_yields_with_shading = [simulate_growth_with_shading(interval, days) for interval in harvest_intervals]

# Plot the results with shading
plt.figure(figsize=(10, 6))
plt.plot(harvest_intervals, avg_yields_with_shading, marker='o', color='r')
plt.xlabel("Harvest Interval (days)")
plt.ylabel("Average Yield per Planted Day with Shading")
plt.title("Average Yield per Planted Day vs Harvest Interval (with Shading Effect)")
plt.show()

# Determine the optimal interval and corresponding yield per planted day with shading
optimal_index = np.argmax(avg_yields_with_shading)
optimal_interval = harvest_intervals[optimal_index]
optimal_yield = avg_yields_with_shading[optimal_index]
print(f"Optimal harvest interval with shading: {optimal_interval} days")
print(f"Corresponding yield per planted day with shading: {optimal_yield}")
```

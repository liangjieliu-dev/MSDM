## Question (a)

We aim to prove that if $ X \sim N(\mu_X, \sigma_X^2) $ and $ Y \sim N(\mu_Y, \sigma_Y^2) $ are independent normal random variables, then their sum $ Z = X + Y $ is also normally distributed, specifically $ Z \sim N(\mu_X + \mu_Y, \sigma_X^2 + \sigma_Y^2) $.

**Proof:**

The cumulative distribution function of $ Z $ is defined as:

$$
F_Z(z) = P(Z \leq z) = P(X + Y \leq z)
$$

Since $ X $ and $ Y $ are independent, their joint probability density function is the product of their individual PDFs:

$$
f(x, y) = f_X(x) f_Y(y)
$$

Thus, we can write:

$$
\begin{align*}
F_Z(z) &= \int_{x + y \leq z} f(x, y) \, dx \, dy \\
&= \int_{-\infty}^{+\infty} \int_{-\infty}^{z - x} f_X(x) f_Y(y) \, dy \, dx \\
&= \int_{-\infty}^{+\infty} f_X(x) \left( \int_{-\infty}^{z - x} f_Y(y) \, dy \right) dx
\end{align*}
$$

To find the PDF $ f_Z(z) $, we differentiate the CDF $ F_Z(z) $ with respect to $ z $:

$$
f_Z(z) = \frac{d}{dz} F_Z(z) = \int_{-\infty}^{+\infty} f_X(x) f_Y(z - x) \, dx
$$

The PDFs of $ X $ and $ Y $ are:

- For $ X $:

  $$
  f_X(x) = \frac{1}{\sqrt{2\pi \sigma_X^2}} \exp\left( -\frac{(x - \mu_X)^2}{2\sigma_X^2} \right)
  $$

- For $ Y $:

  $$
  f_Y(y) = \frac{1}{\sqrt{2\pi \sigma_Y^2}} \exp\left( -\frac{(y - \mu_Y)^2}{2\sigma_Y^2} \right)
  $$

Therefore, the PDF of $ Z $ becomes:

$$
f_Z(z) = \int_{-\infty}^{+\infty} \frac{1}{\sqrt{2\pi \sigma_X^2}} \exp\left( -\frac{(x - \mu_X)^2}{2\sigma_X^2} \right) \cdot \frac{1}{\sqrt{2\pi \sigma_Y^2}} \exp\left( -\frac{(z - x - \mu_Y)^2}{2\sigma_Y^2} \right) dx
$$

First, expand the term $ (z - x - \mu_Y)^2 $:

$$
\begin{align*}
(z - x - \mu_Y)^2 &= \left( (z - \mu_Y) - x \right)^2 \\
&= (z - \mu_Y)^2 - 2(z - \mu_Y)x + x^2
\end{align*}
$$

Now, combine the exponents in the integrand:

$$
\begin{align*}
&-\frac{(x - \mu_X)^2}{2\sigma_X^2} - \frac{(z - x - \mu_Y)^2}{2\sigma_Y^2} \\
&= -\left( \frac{(x - \mu_X)^2}{2\sigma_X^2} + \frac{(z - \mu_Y)^2 - 2(z - \mu_Y)x + x^2}{2\sigma_Y^2} \right) \\
&= -\left( \frac{x^2 - 2\mu_X x + \mu_X^2}{2\sigma_X^2} + \frac{(z - \mu_Y)^2}{2\sigma_Y^2} - \frac{2(z - \mu_Y)x}{2\sigma_Y^2} + \frac{x^2}{2\sigma_Y^2} \right)
\end{align*}
$$

Simplify and group like terms:

$$
-\left( \left( \frac{1}{2\sigma_X^2} + \frac{1}{2\sigma_Y^2} \right) x^2 - \left( \frac{\mu_X}{\sigma_X^2} + \frac{z - \mu_Y}{\sigma_Y^2} \right) x + \frac{\mu_X^2}{2\sigma_X^2} + \frac{(z - \mu_Y)^2}{2\sigma_Y^2} \right)
$$

Let:

- $ A = \frac{1}{2\sigma_X^2} + \frac{1}{2\sigma_Y^2} $
- $ B = \frac{\mu_X}{\sigma_X^2} + \frac{z - \mu_Y}{\sigma_Y^2} $
- $ C = \frac{\mu_X^2}{2\sigma_X^2} + \frac{(z - \mu_Y)^2}{2\sigma_Y^2} $

The exponent becomes:

$$
-\left( A x^2 - B x + C \right) = -\left( A \left( \left( x - \frac{B}{2A} \right)^2 - \left( \frac{B}{2A} \right)^2 \right) + C \right) = -A \left( x - \frac{B}{2A} \right)^2 + \left( \frac{B^2}{4A} - C \right)
$$

Now, the integral becomes:

$$
f_Z(z) = \left( \frac{1}{\sqrt{2\pi \sigma_X^2}} \cdot \frac{1}{\sqrt{2\pi \sigma_Y^2}} \right) \cdot \exp\left( \frac{B^2}{4A} - C \right) \cdot \int_{-\infty}^{+\infty} \exp\left( -A \left( x - \frac{B}{2A} \right)^2 \right) dx
$$

Recognize that:

$$
\int_{-\infty}^{+\infty} \exp\left( -A \left( x - \frac{B}{2A} \right)^2 \right) dx = \sqrt{ \frac{\pi}{A} }
$$

Compute the constant factor:

$$
C_{\text{total}} = \left( \frac{1}{\sqrt{2\pi \sigma_X^2}} \cdot \frac{1}{\sqrt{2\pi \sigma_Y^2}} \right) \cdot \sqrt{ \frac{\pi}{A} } = \frac{1}{2\pi \sqrt{\sigma_X^2 \sigma_Y^2}} \cdot \sqrt{ \frac{\pi}{A} } = \frac{1}{2 \sqrt{\pi A} \sqrt{\sigma_X^2 \sigma_Y^2}}
$$

Compute $ A $:

$$
A = \frac{1}{2\sigma_X^2} + \frac{1}{2\sigma_Y^2} = \frac{\sigma_X^2 + \sigma_Y^2}{2\sigma_X^2 \sigma_Y^2}
$$

Then:

$$
\sqrt{A} = \frac{ \sqrt{ \sigma_X^2 + \sigma_Y^2 } }{ \sqrt{2} \sigma_X \sigma_Y }
$$

So:

$$
C_{\text{total}} = \frac{1}{2\sqrt{\pi} \cdot \frac{ \sqrt{ \sigma_X^2 + \sigma_Y^2 } }{ \sqrt{2} \sigma_X \sigma_Y } \cdot \sigma_X \sigma_Y } = \frac{1}{\sqrt{2\pi (\sigma_X^2 + \sigma_Y^2)}}
$$

Compute $ \frac{B^2}{4A} - C $:

First, note that:

$$
B = \frac{\mu_X}{\sigma_X^2} + \frac{z - \mu_Y}{\sigma_Y^2}
$$

After algebraic manipulation (omitted for brevity), it simplifies to:

$$
\frac{B^2}{4A} - C = -\frac{(z - \mu_X - \mu_Y)^2}{2(\sigma_X^2 + \sigma_Y^2)}
$$

Substitute back into $ f_Z(z) $:

$$
f_Z(z) = \frac{1}{\sqrt{2\pi (\sigma_X^2 + \sigma_Y^2)}} \exp\left( -\frac{(z - \mu_X - \mu_Y)^2}{2(\sigma_X^2 + \sigma_Y^2)} \right)
$$

This is the PDF of a normal distribution with mean $ \mu_Z = \mu_X + \mu_Y $ and variance $ \sigma_Z^2 = \sigma_X^2 + \sigma_Y^2 $.

**Conclusion:**

We have shown that the sum $ Z = X + Y $ of two independent normal random variables $ X $ and $ Y $ is itself normally distributed, with:

- **Mean**: $ \mu_Z = \mu_X + \mu_Y $
- **Variance**: $ \sigma_Z^2 = \sigma_X^2 + \sigma_Y^2 $

Therefore:

$$
Z \sim N(\mu_X + \mu_Y, \sigma_X^2 + \sigma_Y^2)
$$

This completes the proof. Specifically, $Z \sim N(1, 3)$.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

n_samples = 10000  # Number of samples
X = np.random.normal(0, 1, n_samples)
Y = np.random.normal(1, np.sqrt(2), n_samples)

Z = X + Y

mu_Z, sigma_Z = norm.fit(Z)

plt.figure(figsize=(10, 6))

count_x, bins_x, _ = plt.hist(X, bins=50, density=True, alpha=0.4, color='orange', label='Generated X')

count_y, bins_y, _ = plt.hist(Y, bins=50, density=True, alpha=0.4, color='purple', label='Generated Y')

count_z, bins_z, _ = plt.hist(Z, bins=50, density=True, alpha=0.4, color='b', label='Generated Z')

pdf_fitted_z = norm.pdf(bins_z, mu_Z, sigma_Z)
plt.plot(bins_z, pdf_fitted_z, 'r--', label=f'Fitted Normal for Z ($\mu={mu_Z:.2f}, \sigma={sigma_Z:.2f}$)')

pdf_theoretical_z = norm.pdf(bins_z, 1, np.sqrt(3))
plt.plot(bins_z, pdf_theoretical_z, 'g-', label='Theoretical Normal for Z ($\mu=1, \sigma=\sqrt{3}$)')

pdf_theoretical_x = norm.pdf(bins_x, 0, 1)
plt.plot(bins_x, pdf_theoretical_x, 'orange', linestyle='-', label='Theoretical Normal for X ($\mu=0, \sigma=1$)')

pdf_theoretical_y = norm.pdf(bins_y, 1, np.sqrt(2))
plt.plot(bins_y, pdf_theoretical_y, 'purple', linestyle='-', label='Theoretical Normal for Y ($\mu=1, \sigma=\sqrt{2}$)')

plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Histogram and Normal Distribution Fitting for X, Y, and Z')
plt.legend(loc='best')
plt.grid(True)

plt.show()
```

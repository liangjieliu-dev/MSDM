```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

gamma = 1.0
k_B_T = 1.0
dt = 0.01    # Time step
t_max = 1000   # Maximum time
num_steps = int(t_max / dt)

# ========== Task (a): Case with U(x) = 0 ==========

num_particles = 3000
x_a = np.zeros((num_particles, num_steps))

# Plot trajectory of 5 particles by random
for i in range(1, num_steps):
    zeta = np.random.normal(0, np.sqrt(2 * gamma * k_B_T * dt), num_particles)
    x_a[:, i] = x_a[:, i-1] + zeta / gamma

time = np.linspace(0, t_max, num_steps)

for i in range(5):
    plt.plot(time, x_a[i], label=f'Particle {i+1} Trajectory')
plt.xlabel('Time t')
plt.ylabel('Position x(t)')
plt.legend()
plt.title('Trajectory of Brownian particle $U(x) = 0$')
plt.show()

# Plot mean square displacement
mean_square_displacement = np.mean((x_a - x_a[:, 0, np.newaxis])**2, axis=0)

plt.plot(time, mean_square_displacement, label="Simulated result")
plt.plot(time, 2 * k_B_T * time / gamma, label="Theoretical value", linestyle="--")
plt.xlabel('Time t')
plt.ylabel(r'Mean square displacement $\langle [x(t)-x(0)]^2 \rangle$')
plt.legend()
plt.title('Mean square displacement of Brownian motion $U(x) = 0$')
plt.show()

# Compare numerical and theoretical diffusivity
def linear_fit(t, D):
    return 2 * D * t

popt, pcov = curve_fit(linear_fit, time, mean_square_displacement)
numerical_diffusivity = popt[0]
print(f"Numerical diffusivity: {numerical_diffusivity}")

theoretical_diffusivity = k_B_T / gamma
print(f"Theoretical diffusivity: {theoretical_diffusivity}")

diffusivity_error = abs(numerical_diffusivity - theoretical_diffusivity) / theoretical_diffusivity * 100
print(f"Diffusivity error: {diffusivity_error:.2f}%")

# ========== Task (b): Case with U(x) = x^2 / 2 ==========

# Force from potential U(x) = x^2 / 2
def force(x):
    return -x

x_b = np.zeros(num_steps)

for i in range(1, num_steps):
    zeta = np.random.normal(0, np.sqrt(2 * gamma * k_B_T * dt))
    x_b[i] = x_b[i-1] + force(x_b[i-1]) * dt / gamma + zeta / gamma

plt.plot(time, x_b, label=f'Particle Trajectory')
plt.xlabel('Time t')
plt.ylabel('Position x(t)')
plt.legend()
plt.title('Trajectory of Brownian particle $U(x) = x^2 / 2$')
plt.show()

plt.hist(x_b, bins=50, density=True, label="Simulated PDF")

# Theoretical normal distribution
mean = 0
std = np.sqrt(k_B_T / gamma)
x_vals = np.linspace(-5, 5, 100)
pdf_theory = 1 / (std * np.sqrt(2 * np.pi)) * np.exp(-x_vals**2 / (2 * std**2))
plt.plot(x_vals, pdf_theory, label="Theoretical normal distribution", linestyle="--")

plt.xlabel('Position x')
plt.ylabel('Probability density f(x)')
plt.legend()
plt.title('Probability density function in potential $U(x) = x^2 / 2$')
plt.show()

mean_position = np.mean(x_b)
variance = np.var(x_b)

print(f"Mean of f(x): {mean_position}")
print(f"Variance of f(x): {variance}")
```

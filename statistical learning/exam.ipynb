```python
#Q6
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import  confusion_matrix, roc_curve, auc, roc_auc_score
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

Diabetes_train = pd.read_csv('Diabetes_train.csv')
Diabetes_test = pd.read_csv('Diabetes_test.csv')

# Prepare data
X_train = Diabetes_train.drop('diabetes', axis=1)
y_train = (Diabetes_train['diabetes'] == 'neg').astype(int)
X_test = Diabetes_test.drop('diabetes', axis=1)
y_test = (Diabetes_test['diabetes'] == 'neg').astype(int)

# Logistic Regression
log_reg = LogisticRegression(max_iter=10000, solver='liblinear')
log_reg.fit(X_train, y_train)
log_reg_test = log_reg.predict(X_test)
print(confusion_matrix(y_test, log_reg_test))
tn, fp, fn, tp=confusion_matrix(y_test, log_reg_test).ravel()
sensitivity = tp / (tp + fn)
specificity = 1 - fp / (tn + fp)
print(f"      Sensitivity={sensitivity}, Specificity={specificity}\n")

# Predictions
y_probs = log_reg.predict_proba(X_test)[:, 1]

# ROC curve
fpr, tpr, _ = roc_curve(y_test, y_probs)
roc_auc = auc(fpr, tpr)
auc_score = roc_auc_score(y_test, y_probs)
print(f"AUC: {auc_score}")
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC on Test Data by Logistic Regression')
plt.legend(loc="lower right")
plt.show()

########### LDA
import pandas as pd
import numpy as np
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import confusion_matrix, roc_curve, auc
import matplotlib.pyplot as plt
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import train_test_split


Diabetes_train = pd.read_csv('Diabetes_train.csv')
Diabetes_test = pd.read_csv('Diabetes_test.csv')

# Prepare data
X_train = Diabetes_train.drop('diabetes', axis=1)
y_train = (Diabetes_train['diabetes'] == 'neg').astype(int)
X_test = Diabetes_test.drop('diabetes', axis=1)
y_test = (Diabetes_test['diabetes'] == 'neg').astype(int)

# LDA 
lda = LinearDiscriminantAnalysis()
lda.fit(X_train, y_train)
lda_test = lda.predict(X_test)
print(confusion_matrix(y_test, lda_test))
tn, fp, fn, tp=confusion_matrix(y_test, log_reg_test).ravel()
sensitivity = tp / (tp + fn)
specificity = 1 - fp / (tn + fp)
print(f"      Sensitivity={sensitivity}, Specificity={specificity}\n")
# Predictions
y_probs = lda.predict_proba(X_test)[:, 1]

# ROC curve for LDA
fpr, tpr, _ = roc_curve(y_test, y_probs)
roc_auc = auc(fpr, tpr)
auc_score = roc_auc_score(y_test, y_probs)
print(f"AUC: {auc_score}")
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()


############# 
import pandas as pd
import numpy as np
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.metrics import confusion_matrix, roc_curve, auc
import matplotlib.pyplot as plt

Diabetes_train = pd.read_csv('Diabetes_train.csv')
Diabetes_test = pd.read_csv('Diabetes_test.csv')

# Prepare data
X_train = Diabetes_train.drop('diabetes', axis=1)
y_train = (Diabetes_train['diabetes'] == 'neg').astype(int)
X_test = Diabetes_test.drop('diabetes', axis=1)
y_test = (Diabetes_test['diabetes'] == 'neg').astype(int)

# QDA
qda = QuadraticDiscriminantAnalysis()
qda.fit(X_train, y_train)

# Predictions
y_pred = qda.predict(X_test)
y_probs = qda.predict_proba(X_test)[:, 1]

# Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print(conf_matrix)
tn, fp, fn, tp=conf_matrix.ravel()
sensitivity = tp / (tp + fn)
specificity = 1 - fp / (tn + fp)
print(f"      Sensitivity={sensitivity}, Specificity={specificity}\n")
# ROC curve
fpr, tpr, _ = roc_curve(y_test, y_probs)
roc_auc = auc(fpr, tpr)
auc_score = roc_auc_score(y_test, y_probs)
print(f"AUC: {auc_score}")
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```
